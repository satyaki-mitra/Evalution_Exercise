```python
import pandas as pd
```

```python
raw_data = pd.read_csv("data/AAPL-1Min.txt", parse_dates = [['Date', 'Time']])
```

```python
work_data = pd.DataFrame(raw_data).set_index('Date_Time')
work_data.to_pickle('AAPL-1Min.pickle')
```

```python
work_data.dtypes
```

```python
work_data
```

```python
ohlc_dict = {'Open':'mean', 'High':'mean', 'Low':'mean', 'Close': 'mean', 'Volume': 'sum'}
hourly_data = work_data.resample('H').apply(ohlc_dict)
daily_data = work_data.resample('D').apply(ohlc_dict)
```

```python
hourly_report = hourly_data[['Open', 'High', 'Low', 'Close', 'Volume']]
hourly_report.to_pickle('hourly_data.pickle')
hourly_report.describe().to_pickle('hourly_summary.pickle')
```

```python
daily_report = daily_data[['Open', 'High', 'Low', 'Close', 'Volume']]
daily_report.to_pickle('daily_data.pickle')
daily_report.describe().to_pickle('daily_summary.pickle')
```

```python
twominutes_data = work_data.resample('2T').apply(ohlc_dict)
threeminutes_data = work_data.resample('3T').apply(ohlc_dict)
fiveminutes_data = work_data.resample('5T').apply(ohlc_dict)
tenminutes_data = work_data.resample('10T').apply(ohlc_dict)
```

```python
twominutes_data[['Open', 'High', 'Low', 'Close', 'Volume']].to_pickle('AAPL-2Min.pickle')
threeminutes_data[['Open', 'High', 'Low', 'Close', 'Volume']].to_pickle('AAPL-3Min.pickle')
fiveminutes_data[['Open', 'High', 'Low', 'Close', 'Volume']].to_pickle('AAPL-5Min.pickle')
tenminutes_data[['Open', 'High', 'Low', 'Close', 'Volume']].to_pickle('AAPL-10Min.pickle')
```
