# one python program which will create three output csv files.
2. The five output files are exchange.csv, company.csv, company_financials.csv, industry.csv and sector.csv.
3. The data contained in the five csv files are the data for five tables which have a relation.
4. All the data in the supplied csv file should be reflected in the three output files.

```python
import pandas as pd
import numpy as np
```

```python
my_data = pd.read_csv('input_csv/company-data.csv')
```

```python
data = pd.read_csv('input_csv/company-data.csv')
data
```

```python
df = data.rename(columns = {'Co_Code' : 'co_code', 'Co_Name' : 'co_name', 'NSE Symbol' : 'nse_symbol', 'Stock Exchange' : 'exchange_code', 'Industry' : 'industry_name', 'Sector' : 'sector_name', 'Year End' : 'year_end', 'Equity Paid Up' : 'equity_paid_up', 'Net Sales' : 'net_sales', 'PAT' : 'pat', 'PBIDT' : 'pbidt', 'Debt-Equity Ratio[Latest]' : 'debt_equity_ratio_latest', 'PBIDTM (%)[Latest]' : 'pbidtm_percentage_latest'})
df
```

```python
df1 = df[['co_code', 'co_name', 'nse_symbol']].groupby('co_code').first()
df1.to_csv('output_csv/company.csv')
df1
```

```python
df2 = df[['exchange_code']].groupby('exchange_code').first()
df2['exchange_name'] = ['National Stock Exchange', 'National Stock Exchange - Small & Medium Enterprises']
df2.to_csv('output_csv/exchange.csv')
df2
```

```python
df9 = df[['co_code', 'year_end', 'equity_paid_up', 'net_sales', 'pat', 'pbidt', 'debt_equity_ratio_latest', 'pbidtm_percentage_latest' ]].set_index(['co_code', 'year_end'])
df9.to_csv('output_csv/company_financials.csv')
df9
```

```python
df3 = pd.unique(df['sector_name'])
df3 = pd.DataFrame(df3)
df3['sector_code'] = df3.index
df4 = df3.set_index('sector_code')
df5 = df4.rename(columns = {0 : 'sector_name'})
df5.to_csv('output_csv/sector.csv')

df5
```

```python
df6 = pd.unique(df['industry_name'])
df6 = pd.DataFrame(df6)
df6['industry_code'] = df6.index
df7 = df6.set_index('industry_code')
df8 = df7.rename(columns = {0 : 'industry_name'})
df8.to_csv('output_csv/industry.csv')
df8
```
