```python
import json
import pandas as pd
from pprint import pprint
```

```python
%matplotlib inline
```

```python
fp = open("input_csv/ed_trade_data.json")

data = json.load(fp)
pairs = data.keys()
print 'type(pairs) =',type(pairs)

pair_iter = iter(pairs)
```

```python
pair = pair_iter.next()
print 'pair :', pair
trade_data1 = data[pair]
trade_data = pd.DataFrame(trade_data1)
print 'trade data', trade_data.shape
print trade_data.columns
print 'columns types :', trade_data.dtypes
print trade_data.head()
```

```python
trade_data.set_index('date')
trade_data['price'] = trade_data['price'].astype(float)
trade_data['amountBase'] = trade_data['amountBase'].astype(float)
trade_data['amount'] = trade_data['amount'].astype(float)

sdata = trade_data[['date','amount', 'amountBase', 'price']].set_index('date')
sdata
```

```python
sdata['price'].plot()
```

```python
sdata.describe()
```

```python
def remove_outlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    df_out = df_in.loc[(df_in[col_name] < fence_low) | (df_in[col_name] > fence_high)]
    return df_out
remove_outlier(sdata, 'price')
```
