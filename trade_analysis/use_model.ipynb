```python
import json
import pandas as pd
from model import Pair, Trade
```

```python
fp = open("input_csv/ed_trade_data.json")
jsondata = json.load(fp)
pairlist = jsondata.keys()
```

```python
total_data = []
for i in pairlist:
    pair = Pair(i)
    pair_data = jsondata[str(pair)]
    trades = []
    for j in pair_data:
        trade = Trade(pair, j)
        trades.append(trade)
    info_trade = (trades)
    total_data.append(info_trade)
```
