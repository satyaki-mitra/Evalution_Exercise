```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('input_csv/bitcoin-cross-new.csv', sep = ',')
df = data.set_index('#')
df
```

```python
df1 = df.sort_values(by = ['Volume (24h)'], ascending = False)
large_exchanges = df1[['Source', 'Volume (24h)']].head(10)
large_exchanges.to_csv('output_csv/ten_largest_exchanges.csv')
large_exchanges
```

```python
df5 = df[['Source', 'Pair', 'Volume (24h)']]  
# df5 is a slice of the main dataframe(df) containing only 'Source', 'Pair' and 'Volume(24h)' columns.

df6 = df5.groupby(['Source']).sum()           
# df6 is another dataframe which is grouped by 'Source' and contains the total for a group object.

df7 = df6.sort_values(by = ['Volume (24h)'], ascending = False).head(25)
# df7 is another dataframe which contains sorted values of 'Volume (24h)' in descending order.

df7.to_csv('output_csv/top25_exchanges.csv')
df7
```

```python
df3 = df.sort_values(by = ['Price'], ascending = True)
cheapest_pairs = df3[['Pair', 'Price']].head(10)
cheapest_pairs.to_csv('output_csv/ten_cheapest_pairs.csv')
cheapest_pairs
```

```python
df2 = pd.DataFrame(df.Pair.value_counts()).head(10)
df2.to_csv('output_csv/most_traded_pairs.csv')
```
